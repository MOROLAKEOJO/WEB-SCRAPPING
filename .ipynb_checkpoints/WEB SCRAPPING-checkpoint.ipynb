```python

from bs4 import BeautifulSoup as bs
import requests
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


page = 1

t = {"num_of_people_rated":[],
     "rating":[], 
     "discount":[],
     "new_price":[],
     "old_price":[],
     "item name":[]}

while page<=5:
    
    url = "https://www.jumia.com.ng/catalog/?q=shoe&page={}#catalog-listing".format(page)
    html = requests.get(url)
    soup = bs(html.content)

    for i in soup.find_all("a", class_="core"):
        try:
            t["old_price"].append(i.find("div", class_="old").text)
        except:
            t["old_price"].append(np.nan)
        try:
            t["new_price"].append(i.find("div", class_="prc").text)
        except:
            t["new_price"].append(np.nan)
        try:
            t["item name"].append(i.find("h3", class_="name").text)
        except:
            t["item name"].append(np.nan)
        try:
            t["discount"].append(i.find("div", class_="bdg _dsct _sm").text)
        except:
            t["discount"].append(np.nan)
        try:
            t["rating"].append(i.find("div", class_= "stars _s").text)
        except:
            t["rating"].append(np.nan)
        try:
            t["num_of_people_rated"].append(i.find("div", class_= "rev").text)
        except:
            t["num_of_people_rated"].append(np.nan)
    page +=1
```

```python
from selenium import webdriver

from selenium.webdriver.chrome.service import Service as ChromeService 
# part of the selenium library for configuring Chrome browser settings
import requests
from bs4 import BeautifulSoup


t = {"book-title":[],
     "author":[], 
     "price":[],
     "old-price":[],
     "Cover type":[],
     "availability":[]}

page = 1
while page <=50:
    url = "https://www.waterstones.com/books/bestsellers/sort/bestselling/page/{}".format(page)
    path = "C:/Users/USER/OneDrive/Documents/SOFTWARE/chromedriver-win64/chromedriver.exe"

    # Set up the chrome web driver
    chrome_service = ChromeService(executable_path=path)
    driver = webdriver.Chrome(service=chrome_service)

    #Open the webpage
    driver.get(url)

    # wait for the page to load ( you might need to adjust the waiting time)
    #driver.implicitly_wait(10)
    
    
    #get the page source after javascript execution
    page_source = driver.page_source
    soup =BeautifulSoup(page_source)
    
    import pandas as np


    for i in soup.find_all("div", class_="info-wrap"):
        try:
            t["book-title"].append(i.find("div", class_ = "title-wrap").text.strip("\n"))
        except:
            t["book-title"].append(np.nan)
        try:
            t["author"].append(i.find("span", class_="author").text.strip("\n"))
        except:
            t["author"].append(np.nan)
        try:
            t["availability"].append(i.find("span", class_="format").text)
        except:
            t["availability"].append(np.nan)
        try:
            t["price"].append(i.find("span", class_="price").text)
        except:
            t["price"].append(np.nan)
        try:
            t["old-price"].append(i.find("span", class_ = "price-rrp").text)
        except:
            t["old-price"].append("no old price")
        try:
            t["Cover type"].append(i.find("div",class_="book-price").find_all("span")[-1].text.strip())
        except:
            t["cover type"].append(np.nan)
    page +=1
df=pd.DataFrame(t)
#df.to_csv("book.csv", index=False)   
```
